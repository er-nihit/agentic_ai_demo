```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
# Creating a object from WikipediaRetriever.
# Parameters: 
#   top_k_results: Enter to number of desired results
#   lang: Provide the output language (en, hi, fr, etc..) 
retriever = WikipediaRetriever(
    top_k_results=3,
    lang='en'
)
```

```python
# Define your query
query = 'the geopolitical history of india and pakistan from a perspective of a chinese'

# Get results
docs = retriever.invoke(query)
```

```python
docs
```

```python
# Format and print retrieved results
for i, doc in enumerate(docs):
    print(f"\n\n===RESULT {i+1}===")
    print(f"\nCONTENT:\n{doc.metadata['summary']}")
```
