# Langchain Built-in Tools

### DuckDuckGo Search
It is used to query DuckDuckGo Search engine and provide it's output

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()
```

```python
search_value = "Iran vs America"

results = search_tool.invoke(search_value)
print(results)
```

### Shell Tool
It is used to execute shell commands using Langchain

```python
from langchain_community.tools import ShellTool

shell_tool = ShellTool()
```

```python
command = "whoami"

results = shell_tool.invoke(command)
print(results)
```
