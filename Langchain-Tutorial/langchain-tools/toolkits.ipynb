#### Toolkits


```python
from langchain_core.tools import tool
```

```python
@tool
def multiply(a:int, b:int) -> int:
    """Multiplies two numbers"""
    return a*b
```

```python
@tool
def add(a:int, b:int) -> int:
    """Adds two numbers"""
    return a+b
```

```python
# Create a class for the toolkit
class MathToolkit:
    def get_tools(self):
        return [add,multiply]
```

```python
# Creating a toolkit object
toolkit = MathToolkit()
tools = toolkit.get_tools()
```

```python
for tool in tools:
    print(tool.name, ": ", tool.description)
```
