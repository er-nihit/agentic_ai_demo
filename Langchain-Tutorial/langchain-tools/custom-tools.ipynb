# Custom Tools in Langchain

### Using @tool decorator


```python
from langchain_core.tools import tool
```

```python
# Creating a tool
# 1. Create function and add logic you want to perform
# 2. Add hints inside """doc strings""" (Preferred, not mantatory)
# 3. Add type hinting. Provide data type in parameter as well as for return value 
# 4. Add @tool decorator (Preferred, not mantatory)

@tool  # Decorator
def multiply(a: int, b: int) -> int:  # Type hinting
    """Multiply 2 numbers"""  # Doc String
    return a*b
```

```python
# Invokde the tool
# Calling the tool is different from calling a function
# Since it is a runnable, it will perform basic tasks of a runnable
result = multiply.invoke({'a':4, 'b':5})
print(result)
```

```python
# Features provided in all tools
print(multiply.name) # Provides the name of tool, which the function name itself
print(multiply.description) # Provides the description provided in doc string
print(multiply.args) # Expected input data type of tool which is provided by type hinting
```

### Using StructuredTools

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
# Create a class to define the input variables and it's structures
# ... means required=True
class MultiplyInput1(BaseModel):
    a: int = Field(..., description='The first number to multiply')
    b: int = Field(..., description='Second number to multiply')
```

```python
# Defining Tool function
def multiply1(a: int, b: int) -> int:
    return a*b
```

```python
# Creating an object of SructuredTool
# Use function: from_function() to create a tool using a python function
# Parameters:
#   func: function which will be executed when tool is called
#   name: Name of tool.
#   description: Description of Tool
#   args_schema: Input schema required by tool, 
#   It is defined in class earlier by inheriting abstract method of BaseModel

multiply_tool1 = StructuredTool.from_function(
    func=multiply1,
    name="multiply",
    description='Multiplies two numbers',
    args_schema=MultiplyInput1
)
```

```python
# Calling the tool

result = multiply_tool1.invoke({'a':6,'b':8})

print(result)
print(multiply_tool1.name)
print(multiply_tool1.description)
print(multiply_tool1.args)
```

### Using BaseTool Class

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
# Defining a input schema for tool (same as StructuredTool) using Pydantic Schema
# '...' means required=True in new version of Pydantic
class MultiplyInput2(BaseModel):
    a: int = Field(..., description='The first number to multiply')
    b: int = Field(..., description='The second number to multiply')
```

```python
# Defining a tool by inheriting BaseTool Class
class MultiplyTool2(BaseTool):
    name: str = "multiply" # Providing a name to the tool
    description: str = "Multiplies 2 numbers" # Providing description of tool
    args_schema: Type[BaseModel] = MultiplyInput2 # Defining input schema within the tool

    # Defining the function which will be perfoemed when the tool is called
    # Always define the task within the _run() function

    def _run(self, a: int, b: int) -> int:
        return a*b
```

```python
# Now creating a object of the tool and calling it
multiply2 = MultiplyTool2()

result = multiply2.invoke({'a':4,'b':9})

print(result)
print(multiply2.name)
print(multiply2.description)
print(multiply2.args)
```
