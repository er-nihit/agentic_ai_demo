Task: Simple task to take user input and provide LLM based answer  

**States**:
- `question`: User input
- `answer`: LLM Generation

```python
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatOpenAI()
```

```python
class LLMState(TypedDict):
    question: str
    answer: str
```

```python
def ask_llm(state: LLMState) -> LLMState:

    state['answer'] = model.invoke(state['question']).content

    return state
```

```python
graph = StateGraph(LLMState)

graph.add_node('ask_llm', ask_llm)

graph.add_edge(START, 'ask_llm')
graph.add_edge('ask_llm',END)

workflow = graph.compile()
```

```python
workflow.invoke({'question':'What is capital of India'})
```
