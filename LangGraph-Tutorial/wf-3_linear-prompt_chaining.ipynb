**Task**: Create a outline and blog based on topic provided by user.  

**States**:
- `topic`: user input
- `outline`: LLM Generated
- `blog`: LLM Generated

```python
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatOpenAI(model='gpt-5-nano')
```

```python
class BlogState(TypedDict):
    topic: str
    outline: str
    blog: str
```

```python
def create_outline(state: BlogState) -> BlogState:
    topic = state['topic']

    prompt = f"Create a outline for a blog on {topic}"

    result = model.invoke(prompt)

    state['outline'] = result.content

    return state
```

```python
def create_blog(state: BlogState) -> BlogState:
    outline = state['outline']

    prompt = f"Create a blog on the mentioned outline  \n {outline}"

    result = model.invoke(prompt)

    state['blog'] = result.content

    return state
```

```python
graph = StateGraph(BlogState)

graph.add_node('create_outline', create_outline)
graph.add_node('create_blog', create_blog)

graph.add_edge(START, 'create_outline')
graph.add_edge('create_outline', 'create_blog')
graph.add_edge('create_blog', END)

workflow = graph.compile()
```

```python
initial_state = {'topic': 'Recession'}

final_state = workflow.invoke(initial_state)
```
