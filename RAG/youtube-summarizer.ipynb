### 1. Indexing

```python
# Get youtube link/ID from the user
# Video ID is the value between 'v=' and '?' in the link
video_link = "https://www.youtube.com/watch?v=0wPe2pkwVfY"
video_id = "LPZh9BOjkQs"


try:
    # This block will check if the transcripts are enabled for the video
    transcript_list = YouTubeTranscriptApi().fetch(video_id)

    # Merge the automatically generated chunks into a single string
    transcript = " ".join(chunk['text'] for chunk in transcript_list)
    print(transcript)

except:
    # If the video does not have captions enabled, it will throw this error
    print("No captions enabled for this video")
```
